# How to Design Professional Venn Diagrams in Python

A Venn diagram is an illustration used to depict logical relationships between different groups or sets.

Each group is represented using a circle. The size of each circle corresponds to the size/importance of the group.

```python
#Import libraries
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
%matplotlib inline
```

An Example code provided by the [matplotlib-venn](https://github.com/konstantint/matplotlib-venn) library.

```python
# libraries
from matplotlib import pyplot as plt
import numpy as np
from matplotlib_venn import venn3, venn3_circles
 
# Make a Basic Venn
v = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels = ('A', 'B', 'C'))
 
# Custom it
v.get_patch_by_id('100').set_alpha(1.0)
v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('100').set_text('Unknown')

v.get_label_by_id('A').set_text('Set "A"')
v.get_label_by_id('B').set_text('Set "B"')
v.get_label_by_id('C').set_text('Set "C"')


c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')

c[0].set_lw(1.0)
c[0].set_ls('dotted')
 
# Add title and annotation

plt.title("A Cool Venn diagram")
plt.annotate('Unknown set', xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'));
plt.savefig('sample_venn.png', dpi=800)
```

Defining the subjects: **English** and **French**

```python
English = {'John', 'Amy', 'Howard', 'Lucy', 'Alice', 'George', 'Jacob', 'Rajesh', 'Remy', 'Tom'}
French = {'Arthur', 'Leonard', 'Karan', 'Debby', 'Bernadette', 'Alice', 'Ron', 'Penny', 'Sheldon', 'John'}
```

We will plot our Venn Diagram using the `venn2` function of matplotlib-venn:

```python
venn2([English,French]);
# plt.savefig('venn_2.png',dpi = 200)
```

We can also add labels to the plot and individual circles.

```python
# adding labels using set_labels parameter
venn2([English, French], set_labels = ('English Class', 'French Class'));

# plot title
plt.title('Relationship b/w students taking English and French Class\n');
```

Defining the subjects: **Math** and **Science**

```python
Math = {'Amy','Leo','Ash','Brandon','Sara','Alice','Ron','Annie','Jake','Zac'}
Science = {'Dexter','Geeta', 'James', 'Charles','Sammy','Brandon','Aayush','Josephy','Casey','Will'}
```

We will now plot a Venn diagram with three sets: **English**, **French**, and **Maths**.  
We do this using the `venn3` function:

```python
# use venn3 when you have three groups
venn3([English, French, Math], set_labels = ('English Class', 'French Class','Maths Class'));

# plot title
plt.title('Relationship b/w students taking English, French and MathsClass\n');
```

Now, let us consider using Science as the third set rather than Math.

```python
# using Science instead of Math
venn3([English,French,Science], set_labels = ('English', 'French','Science'));

# plot title
plt.title('Relationship b/w students taking English, French and Science Class\n');
```

Mandatory Subject **Fire Safety**

```python
# Taking a union of all sets
fire_safety= English | French | Math | Science

#Plotting alongside English and French
venn3([English,French,fire_safety], set_labels = ('English', 'French','Fire-Safety'));

plt.title('Relationship b/w students taking English, French and Fire-Safety Class\n');
```

## Using Id for Customization

You can customize the **colors of each area** of the diagram with the `get_patch_by_id` method. 

We can also customize the **text for each region** using the `get_label_by_id` method.

```python
# defining our venn
v = venn2(subsets=(1,1,1), set_labels = ('Set A', 'Set B'))

# setting color of id 11 to purple
v.get_patch_by_id('11').set_color('purple')
#setting label of id 10 to New Text
v.get_label_by_id('10').set_text('New Text')

#plot title
plt.title('Customizing Venn using IDs');
```
